In this mini project we will create a chessboard alternate color pattern using matplotlib and Numpy it will explain how Matplotlib works 

```python
#we will start by importing libraries
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LogNorm
```

```python
#Now we will create axis  using Numpy
dx, dy = 0.015, 0.05
x = np.arange(-4.0, 4.0, dx)
y = np.arange(-4.0, 4.0, dy)
X, Y = np.meshgrid(x, y)
extent = np.min(x), np.max(x), np.min(y), np.max(y)
z1 = np.add.outer(range(8), range(8)) % 2
plt.imshow(z1, cmap="binary_r" , interpolation="nearest" , extent=extent, alpha= 1)
#now we will define this as chessboard
def chess(x, y):
    return (1 - x / 2 + x ** 5 + y ** 6) * np.exp(-(x ** 2 + y ** 2))
z2 = chess(X, Y)
plt.imshow(z2, alpha=0.7, interpolation="bilinear", extent=extent)
plt.title("Chess Board with Python")
plt.show()
```

So this is how we can create and visualize a chessboard with code.
