```python
#Importing Libraries
import numpy as np
import pandas as pd
```

```python
#Reading CASE-SHILLER Index into a dataframe
df_CS = pd.read_csv("CSUSHPISA.csv")

#Changing dtype of date column
df_CS["DATE"] = pd.to_datetime(df_CS["DATE"])

#Resetting Index
df_CS.reset_index(inplace = True)
df_CS.drop(columns = ["index"], inplace = True)

# Creating "Year" and "Month" columns
df_CS["Year"] = pd.DatetimeIndex(df_CS["DATE"]).year
df_CS["Month"] = pd.DatetimeIndex(df_CS["DATE"]).month
print(df_CS.shape)
df_CS.head()
```

```python
# Reading Unemployment Rate Data into a dataframe
df_unemp = pd.read_csv("UNRATE.csv")
#as numbers of columns are more in this data so we are gonna drop some 
df_unemp.drop([261,262], inplace = True)
print(df_unemp.shape)
df_unemp.tail()
```

```python
# Interest Rate Data
df_Fed_rate = pd.read_csv("FEDFUNDS (2).csv").drop([261,262])
print(df_Fed_rate.shape)
df_Fed_rate.head(10)
```

```python
# Reading Per Capita GDP Data into a dataframe
df_GDP = pd.read_csv("GDP_PER_CAPITA.csv",names = ["DATE", "Per_Capita_GDP"], skiprows = 1)
print(df_GDP.shape)
df_GDP.head(10)
#This is Quarterly Data we will use linear interpolation after final dataframe is created 
```

```python
#
df_COMMODITY_PRICES = pd.read_csv("COMMODITY_PRICES.csv", names = ["DATE", "Cons_Materials"], skiprows = 1)
df_COMMODITY_PRICES.drop([261], inplace = True)
print(df_COMMODITY_PRICES.shape)
df_COMMODITY_PRICES.tail()
```

```python
#CPI 
df_CPI = pd.read_csv("CPI.csv", names = ["DATE", "CPI"], skiprows = 1).drop([261])
print(df_CPI.shape)
df_CPI.tail()
```

```python
# Monthly new house supply
df_house = pd.read_csv("PVT_HOUSE_CONSTRUCTED.csv", names = ["DATE", "Houses"], skiprows = 1).drop([261])
print(df_house.shape)
df_house.tail(10)
```

```python
# Population above 65

df_oldpop = pd.read_csv("oldpop.csv", names = ["DATE", "old_percent"], skiprows = 1)
print(df_oldpop.shape)
df_oldpop.tail()
```

```python
# Urban Population Percent

df_urban = pd.read_csv("urbanpop2.csv").drop([261])
print(df_urban.shape)
df_urban.tail()
```

```python
# Housing Subsidies

df_subsidy = pd.read_csv("Subsidies.csv", names = ["DATE", "Subsidy"], skiprows = 1)
print(df_subsidy.shape)
df_subsidy.tail()
```

```python
# Working age population

df_working = pd.read_csv("workingpop.csv", names = ["DATE", "Working_Population"], skiprows = 1).drop([261])
print(df_working.shape)
df_working.tail()
```

```python
# Real Average Household Income

df_income = pd.read_csv("DSPIC96.csv", names = ["DATE", "Income"], skiprows = 1).drop([261])
print(df_income.shape)
df_income.tail()
```

```python
# Number of households

df_households = pd.read_csv("TOTAL_HOUSE.csv", names = ["DATE", "Num_Households"], skiprows = 1)
print(df_households.shape)
df_households.tail()
```

```python
# Merging Per Capita GDP (Quarterly data)
df_GDP["DATE"] = pd.to_datetime(df_GDP["DATE"])
df_CS = pd.merge(df_CS,df_GDP, how = "left")
df_CS.head()
```

```python
# Concating dataframes having monthly data to create one dataframe
df = pd.DataFrame()
df_bymonth = [df_CS, df_working, df_house, df_CPI, df_unemp, df_COMMODITY_PRICES, df_Fed_rate]
for df1 in df_bymonth:
    df1["DATE"] = pd.to_datetime(df1["DATE"])
    df1 = df1.set_index("DATE")
    df = pd.concat([df,df1], axis = 1)
print(df.shape)
df.head()
```

```python
# Merging other dataframes 
others = [df_urban, df_households, df_income, df_subsidy, df_oldpop]
for df1 in others:
    if "Year" not in df1.columns:
        df1["Year"] = pd.DatetimeIndex(df1["DATE"]).year
        df1.set_index("DATE", inplace = True)
        df = pd.merge(df, df1, how = "left", on = "Year")
    else:
        df1.set_index("DATE", inplace = True)
        df = pd.merge(df, df1, how = "left", on = "Year")
df["DATE"] = df_CS["DATE"]
df.set_index("DATE", inplace = True)
df.head()
```

```python
df.isna().sum()
```

```python
# Filling missing values in the Per_Capita_GDP column using linear interpolation
df["Per_Capita_GDP"] = df["Per_Capita_GDP"].interpolate()
```

```python
df.head()
```

```python
df.dropna(inplace = True)
```

```python
df.to_csv("prepared_dataset.csv")
```
